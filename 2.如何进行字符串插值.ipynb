在不导入Template类的情况下，有3种方法进行字符串插值。

字符串插值是将字符串中的占位符替换为局域变量的过程。

字符串插值，就是将变量、常量以及表达式插入到字符串中的一种技术，它可以避免字符串拼接的问题，

```python
    #Python 3.6中提供，即f-string，字面量格式化字符串。

name = "吉多·范罗苏姆"
print(f"姓名：{name}")
```
